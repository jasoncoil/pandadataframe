```python
import pandas as pd
import re
```

```python
#Opening Text File and Reading it
text_file = open("marcRebillet.txt", "r")
lines = text_file.read()

#Splitting it into an array of one words, and making a list
Array = re.split(' |\n',lines)

#Creating a data frame and counting the unique values within it
df = pd.DataFrame({'Unique Marc Words':Array})
df['Frequency'] = df.groupby('Unique Marc Words')['Unique Marc Words'].transform('count')

#Dropping the duplicate values and only keepng one of them
df = df.drop_duplicates(subset=['Unique Marc Words', 'Frequency'])

#Locating the words with a frequency higher than 20
df.loc[(df['Frequency'] > 40)]

#Creating a new dataframe with only words said more than 40 times
#and ordering them from highest frequency to lowest frequency
df40 = df.loc[(df['Frequency'] > 40)]
df40.sort_values(['Frequency'], ascending = False)

#NEXT STEPS: 
#1. Change all capital letters to lowercase and then adjust accordingly
#Examples: the word 'oh' and 'Oh' are different, one has frequency = 71 and other frequency = 20.. need to add them
#2. Rearrange in order of highest frequency to lowest
#3.Maybe make an infographic or graph/chart and possibly only deal with frequency above a certain threshold
```
